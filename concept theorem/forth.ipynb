# 클래스

```python
import abc
#볼륨 클래스

class Volume(metaclass = abc.ABCMeta): #추상 클래스
    vol = 20
    
    @abc.abstractmethod  # 추상 메서드로 지정
    def volumeUp(self, vol):
        pass

    @abc.abstractmethod 
    def volumeDown(self, vol):
        pass
        
############################################

class TVV(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        if self.vol >100:
            print("사용할 수 없는 방식입니다.")
            self.vol-=vol        
        else :
            print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        
    def volumeDown(self, vol):
        self.vol -= vol
        if self.vol <0:
            print("사용할 수 없는 방식입니다.")
            self.vol+=vol            
        else :
            print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))            

############################################
class RadioV(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        if self.vol >100:
            print("사용할 수 없는 방식입니다.")
            self.vol-=vol
        else :
            print("Radio 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))            
        
    def volumeDown(self, vol):
        self.vol -= vol
        if self.vol <0:
            print("사용할 수 없는 방식입니다.")
            self.vol+=vol   
        else :
            print("Radio 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))            

############################################
class SpeakerV(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        if self.vol >100:
            print("사용할 수 없는 방식입니다.")
            self.vol+=vol            
        else :
            print("Speaker 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))            
        
    def volumeDown(self, vol):
        self.vol -= vol
        if self.vol <0:
            print("사용할 수 없는 방식입니다.")
            self.vol+=vol  
        else :
            print("Speaker 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
            

#############################################
```
